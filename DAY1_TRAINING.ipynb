```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
class LinkedList:
  def __init__(self):
    self.head=None
  def insertAtBegin(self,new_data):
    new_node=Node(new_data)
    new_node.next=self.head
    self.head=new_node

  def printList(self):
    temp=self.head
    while temp:
      print(f'{temp.data}',end=" ")
      temp = temp.next

L1=LinkedList()
L1.insertAtBegin(10)
L1.insertAtBegin(20)
L1.insertAtBegin(44)
L1.insertAtBegin(50)
L1.insertAtBegin(60)
L1.printList()
```

# INSERT_AT_END

```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
class LinkedList:
  def __init__(self):
    self.head=None
  def insertAtEnd(self,new_data):
    new_node=Node(new_data)
    if self.head is None:
      self.head=new_node
      return
    temp=self.head
    while temp.next:
      temp=temp.next
    temp.next=new_node

  def printList(self):
    temp=self.head
    while temp:
      print(f'{temp.data}',end=" ")
      temp = temp.next

L1=LinkedList()
L1.insertAtEnd(10)
L1.insertAtEnd(20)
L1.insertAtEnd(44)
L1.insertAtEnd(50)
L1.insertAtEnd(60)
L1.printList()
```

# INSERT_AT_MIDDLE

```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
class LinkedList:
  def __init__(self):
    self.head=None
  def insertAtEnd(self,new_data):
    new_node=Node(new_data)
    if self.head is None:
      self.head=new_node
      return
    temp=self.head
    while temp.next:
      temp=temp.next
    temp.next=new_node

  def printList(self):
    temp=self.head
    while temp:
      print(f'{temp.data}',end=" ")
      temp = temp.next

  def insertAtmiddle(self,prev_node,new_data):
    if self.head is None:
      print("Previous node cannot be None.")
    new_node=Node(new_data)
    new_node.next=prev_node.next # The new node points to what the previous node was pointing to
    prev_node.next=new_node     # The previous node now points to the new node

L1=LinkedList()
L1.insertAtEnd(2)
L1.insertAtEnd(4)
L1.insertAtEnd(5)
L1.insertAtEnd(10)
L1.printList()
print()

L1.insertAtmiddle(L1.head.next,12)
L1.printList()
print()

L1.insertAtmiddle(L1.head.next,1)
L1.printList()
print()
```

# TO DETECT THE CYCLE

```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
class LinkedList:
  def __init__(self):
    self.head=None
  def insertAtEnd(self,new_data):
    new_node=Node(new_data)
    if self.head is None:
      self.head=new_node
      return
    temp=self.head
    while temp.next:
      temp=temp.next
    temp.next=new_node

  def printList(self):
    temp=self.head
    while temp:
      print(f'{temp.data}',end=" ")
      temp = temp.next

  # detect_cycle should be a method of the LinkedList class
  def detect_cycle(self):
    fast=self.head
    slow=self.head
    # The loop condition was incomplete; it needs to iterate and move pointers
    while fast and fast.next:
      slow = slow.next
      fast = fast.next.next
      if fast==slow:
        print("Cycle detected")
        return True
    print("No cycle detected")
    return False

L1=LinkedList()
L1.insertAtEnd(10)
L1.insertAtEnd(20)
L1.insertAtEnd(44)
L1.insertAtEnd(50)
L1.insertAtEnd(60)
L1.printList()
print()
L1.detect_cycle()

# Example of creating a cycle for testing
# L1.head.next.next.next.next.next = L1.head.next # Creates a cycle at node 20
# L1.detect_cycle()
```

```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
class LinkedList:
  def __init__(self):
    self.head=None
  def insertAtBegin(self,new_data):
    new_node=Node(new_data)
    new_node.next=self.head
    self.head=new_node

  def printList(self):
    temp=self.head
    while temp:
      print(f'{temp.data}',end=" ")
      temp = temp.next
    print()
  def counting(self):
    count=0
    temp=self.head
    while temp:
      count+=1
      temp=temp.next
    return count
L1=LinkedList()
L1.insertAtBegin(10)
L1.insertAtBegin(20)
L1.insertAtBegin(44)
L1.insertAtBegin(50)
L1.insertAtBegin(60)
L1.printList()
print("Number of nodes:", L1.counting())
```
